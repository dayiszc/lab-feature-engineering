```python
numerical = df.select_dtypes(include=[np.number])  #checking categoricals and numerical values
categorical = df.select_dtypes(include=[object])
print(numerical)
print(categorical)
```

```python
numerical_nulls_percent_df = pd.DataFrame(numerical.isna().sum()/len(numerical)).reset_index()  #checking percentage of nulls in numericals
numerical_nulls_percent_df.columns = ['column_name', 'nulls_percentage']

numerical_nulls_percent_df
```

```python
categorical_nulls_percent_df = pd.DataFrame(numerical.isna().sum()/len(numerical)).reset_index()  #checking percentage of nulls in categoricals
categorical_nulls_percent_df.columns = ['column_name', 'nulls_percentage']

categorical_nulls_percent_df
```

```python
numerical['WEALTH1'].value_counts(dropna = False)
#Nan values 44732
```

```python
sns.histplot(numerical[numerical['WEALTH1'].notna()]['WEALTH1'])
plt.show()
```

```python
mode = numerical['WEALTH1'].mode()[0]
numerical['WEALTH1'] = numerical['WEALTH1'].fillna(mode)
```

```python
numerical['ADI'].isna().sum()
```

```python
fig, ax = plt.subplots(1,3,figsize=(16,5))
sns.histplot(numerical['ADI'], ax = ax[0])
sns.histplot(numerical["ADI"].fillna(np.mean(numerical[numerical['ADI'].notna()]['ADI'])), ax = ax[1])
sns.histplot(numerical['ADI'].fillna(np.median(numerical[numerical['ADI'].notna()]['ADI'])),ax = ax[2])
plt.show()
```

```python
numerical['ADI'] = numerical["ADI"].fillna(np.median(numerical['ADI']))
numerical['DMA'] = numerical["DMA"].fillna(np.median(numerical['DMA']))
numerical['MSA'] = numerical["MSA"].fillna(np.median(numerical['MSA']))

print(numerical['ADI'])
print(numerical['DMA'])
print(numerical['MSA'])
```
